```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
datalink = '/content/drive/MyDrive/0.Samsung CSR AIML/Resources/SIC-Artificial-Intelligence/SIC_AI_Coding_Exercises/SIC_AI_Chapter_05_Coding_Exercises/Clustering_iris.csv'
df = pd.read_csv("/content/Clustering_iris.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cluster_df = df.drop('variety', axis=1)

cluster_df.head()
```

# KMeans

```python
cluster = KMeans(n_clusters=3, random_state=0)

cluster.fit(cluster_df)
```

```python
cluster.fit_predict(cluster_df)
```

```python
df['cluster'] = cluster.fit_predict(cluster_df)

df.head()
```

```python
df['cluster'].value_counts()
```

```python
df['cluster'].value_counts().hist()
```

```python
import seaborn as sns

sns.pairplot(df, hue='cluster',palette='tab10')
```

```python
sns.boxplot(x=df['sepal.width'],y=df['cluster'])
```

```python
sns.heatmap(cluster_df.corr(), annot=True)
```

```python
# help(sns)
```

```python
sns.scatterplot(x=df['petal.length'],
            y=df['petal.width'],
            hue=df['cluster'],
            palette='tab10')
```

# Heirarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hcluster = AgglomerativeClustering(n_clusters=3)
```

```python
hcluster.fit_predict(cluster_df)
```

```python
df['hcluster'] = hcluster.fit_predict(cluster_df)

df.sample(2)
```

```python
df['hcluster'].value_counts()
```

```python
sns.scatterplot(x=df['petal.length'],
            y=df['petal.width'],
            hue=df['hcluster'],
            palette='tab10')
```

# DBScan

```python
from sklearn.cluster import DBSCAN
```

```python
dcluster = DBSCAN(eps=0.5, min_samples=5)
```

```python
dcluster.fit_predict(cluster_df)
```

```python
df['dcluster'] = dcluster.fit_predict(cluster_df)

df.sample(2)
```

```python
df['dcluster'].value_counts()
```

```python
sns.scatterplot(x=df['petal.length'],
            y=df['petal.width'],
            hue=df['dcluster'],
            palette='tab10')
```
