```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
X = df.data

y = df.target
```

```python
X
```

```python
from sklearn.decomposition import PCA

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# Standardizing the data
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Applying PCA
pca_iris = PCA()
pca_iris.fit(X_scaled)
```

```python
# Explained variance ratio
explained_variance_X = pca_iris.explained_variance_ratio_
```

```python
explained_variance_X
```

```python
# Showing the components (loadings)
components_X = pd.DataFrame(pca_iris.components_,columns=df.feature_names)
```

```python
components_X
```

Classify the flowers

```python
from sklearn.ensemble import RandomForestClassifier

# Fitting the Random Forest model
rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(X, y)
```

```python
rf_classifier.feature_importances_
```

```python
importances_iris =rf_classifier.feature_importances_
```

```python
# Creating a DataFrame for feature importances
feature_importance_iris_df = pd.DataFrame({
    'Feature': df.feature_names,
    'Importance': importances_iris
}).sort_values(by='Importance', ascending=False)
```

```python
feature_importance_iris_df
```

```python
import seaborn as sns
sns.barplot(x=feature_importance_iris_df['Importance'],y=feature_importance_iris_df['Feature'])
```
